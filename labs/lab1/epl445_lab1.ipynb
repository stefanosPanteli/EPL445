````python
# Read RGB color image
img =cv2.imread('/content/messi.jpg')

# Show RGB image
print('RGB image')
cv2_imshow(img)

# Read grayscale image
img_gray =cv2.imread('/content/messi.jpg',0)

# Show grayscale image
print('Grayscale image')
cv2_imshow(img_gray)
````

````python
# Print image info
print ('Image shape (height, width, channels): {}'.format(img.shape))
print ('Image size (#pixels): {}'.format(img.size))
print ('Image data type: {}'.format(img.dtype))
````

````python
# Image ROI
# 1. Copy the image to another variable
# 2. Select the ROI (ball)
# 3. Copy the ROI to another region in the image
img2 = img.copy()
ball = img2[280:340, 330:390]
img2[273:333, 100:160] = ball

# Show the new image
cv2_imshow(img2)
````

````python
# Split image to RGB channels
b,g,r = cv2.split(img)
print('Red channel')
cv2_imshow(r) # Red channel
print('Green channel')
cv2_imshow(g) # Green channel
print('Blue channel')
cv2_imshow(b) # Blue channel
````

````python
# Show red channel in chromatic tone
img3 = img.copy()
img3[:,:,0]=0 # img3[height,width,depth] Zeroing blue channel
img3[:,:,1]=0 # Zeroing Green channel
img_red = cv2.merge((img3[:,:,0],img3[:,:,1],r))
print('Red channel in chromatic tone')
cv2_imshow(img_red)
````

````python
# Convert and split image to YCrCb channels
img4 = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
Y, Cr, Cb = cv2.split(img4)
print('Luminance')
cv2_imshow(Y) # Luminance
print('Chroma red')
cv2_imshow(Cr) # Chroma red
print('Chroma blue')
cv2_imshow(Cb) # Chroma blue
````

````python
# Save image in different formats
cv2.imwrite('image.jpg', img_red)
cv2.imwrite('image.png', img_red)
cv2.imwrite('image.tiff', img_red)
````

## Histogram

You can consider a histogram as a graph that gives you a general idea about the distribution of image intensity. It is a graph with pixel values ranging from 0 to 255 (not always) in X-axis and a corresponding number of pixels in the Y-axis image.

For grayscale images:

*   0 represents black color
*   255 represents white color

histogram_sample.jpg


````python
# Plot the histogram of the grayscale image using matplotlib library
plt.hist(img_gray.ravel(),256,[0,256]); # Arrange image pixels in numerical order by defining the number of elements and the range of the pixels

# Show the histogram
plt.show()
````
